### 개봉연도 데이터 정제하기(데이터 전처리, Preprocessing)

### genres 분석

```python
movies['genres']
```

```python
sample_genre = movies['genres'][1]
sample_genre
```

```python
sample_genre.split("|")
```

```python
genres_list = list(movies['genres'].apply(lambda x: x.split("|")))
genres_list[:3]
```

```python
flat_list = []
for sublist in genres_list:
    for item in sublist:
        flat_list.append(item)
        
flat_list
```

```python
set(flat_list)
```

```python
genres_unique = list(set(flat_list))
```

```python
len(genres_unique) # 19개의 장르가 있음 한개는 no genres listed
```

## 장르 데이터 숫자형으로 변환하기

```python
'Adventure' in sample_genre
```

```python
'Horror' in sample_genre
```

```python
movies['Comedy']=movies['genres'].apply(lambda x: 'Comedy' in x)
```

```python
movies
```

```python
genres_dummies = movies['genres'].str.get_dummies(sep='|')
```

```python
# pickle로 저장
genres_dummies.to_csv("D:/추천알고리즘강의/data/ml-latest-small/genres.csv")
```

```python
#두장르의 관계가 1에 가깝다는 것은 두 장르가 같이 출현한다는 것
#두장르의 관계가 -1에 가깝다는 것은 두 장르가 아주 드물게 출현하거나, 겹치는 영역이 없음
#두장르의 관계가 0에 가까운건 별 관계가 없다는 것.
genres_dummies.corr()
```

```python
plt.figure(figsize=(30, 30))
sns.heatmap(genres_dummies.corr(), annot=True)
```
