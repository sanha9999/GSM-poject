```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## Read Data

```python
movies = pd.read_csv('D:/추천알고리즘강의/data\ml-latest-small/movies.csv')
ratings = pd.read_csv('D:/추천알고리즘강의/data/ml-latest-small/ratings.csv')
genres = pd.read_csv('D:/추천알고리즘강의/data/ml-latest-small/genres.csv')
```

```python
ratings.tail()
```

```python
genres.sample()
```

## Preprocessing

```python
ratings = ratings.merge(genres, left_on='movieId', right_index=True)
ratings.sample()
```

```python
ratings = ratings.replace(0, np.nan)
ratings.sample()
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(ratings, random_state=99, test_size=.1)
```

```python
print(train.shape)
print(test.shape)
```

## Item profile

```python
genres
```

## User profile

```python
genres_cols = genres.columns
genres_cols
```

```python
for cols in genres_cols:
    train[cols]=train[cols] * train['rating']
```

```python
train.head()
```

```python
#각 유저별로 Action 아이템에 준 평점의 평균
#train.groupby('userId')['Action'].mean()
user_profile = train.groupby('userId')[genres_cols].mean()
user_profile.sample()
```
